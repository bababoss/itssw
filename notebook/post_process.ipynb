```python
import glob ,os,json,requests
import numpy as np
import uuid
from matplotlib.pyplot import imshow
from PIL import Image
from shutil import copyfile
%matplotlib inline

# im_path="/home/cogknit/Documents/datasets/labelled_data/data/set12/"
# label_path="/home/cogknit/Documents/datasets/labelled_data/data/set12/001/"


def prepare_dataset(im_path,label_path):
    final_image_path="/home/gpu-machine/DATASET_CLEAN/rnpr_image_dataset_25_aug/"
    final_label_path="/home/gpu-machine/DATASET_CLEAN/rnpr_label_dataset_25_aug/"

    image_list=glob.glob(im_path+"*.jpg")
    bbx_list=glob.glob(label_path+"*.txt")
    label_list=glob.glob(label_path+"*.label")
    print(len(image_list),len(bbx_list),len(label_list))

    count=0
    for i in range(len(label_list)):
        filename=label_list[i].split('.')[0].split('/')[-1]
        with open(label_list[i],'r') as f:
            label=f.readlines()
        with open(label_path+filename+'.txt','r') as f:
            bbx=f.readlines()
            #print(bbx[2].strip())
        if os.path.exists(im_path+filename+'.jpg'):
            pil_im = Image.open(im_path+filename+'.jpg', 'r')
            #imshow(np.asarray(pil_im))
            uid=uuid.uuid4().hex
            st="Station_2_New_"
            save_im_path=final_image_path+im_path.split('/')[-2]+st+uid+".jpg"
            save_car_bbx_path=final_label_path+"car_bbx/"+im_path.split('/')[-2]+st+uid+".txt"
            save_np_bbx_path=final_label_path+"np_bbx/"+im_path.split('/')[-2]+st+uid+".txt"
            save_label_path=final_label_path+"np_text_lable/"+im_path.split('/')[-2]+st+uid+".label"
            #print(save_im_path,save_car_bbx_path)

            if label[0].strip() and bbx[0].strip()=='2' and len(label[0].strip())<=10: 
                print(label[0].strip()," ",bbx[1].strip())
                with open(save_car_bbx_path,'w') as f:
                    f.write("1\n"+bbx[1].strip()+"\n")
                with open(save_np_bbx_path,'w') as f:
                    f.write("1\n"+bbx[2].strip()+"\n")
                #copyfile(label_path+filename+'.txt', save_bbx_path)
                copyfile(label_list[i], save_label_path)            
                pil_im.save(save_im_path, "JPEG", quality=100, optimize=True, progressive=True)
                count+=1

    print(count)
    return count
```

```python
final_dataset_history={}
```

```python
im_path="/home/gpu-machine/dataset_new/Station_2_New/4/"
label_path="/home/gpu-machine/dataset_new/Station_2_New/4/001/"
cont=prepare_dataset(im_path,label_path)
final_dataset_history[im_path]=cont
```
