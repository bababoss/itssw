```python
import logging
import logging.handlers as handlers
import time

logger = logging.getLogger('file_name')
logger.setLevel(logging.INFO)

# Here we define our formatter
formatter = logging.Formatter('%(asctime)s,%(name)s,%(levelname)s,%(message)s')

# ACCESS LOGS
logHandler = handlers.RotatingFileHandler('access.log', maxBytes=500000, backupCount=2)
logHandler.setLevel(logging.INFO)
logHandler.setFormatter(formatter)

# ERROR LOGS
errorLogHandler = handlers.RotatingFileHandler('error.log', maxBytes=500000, backupCount=2)
errorLogHandler.setLevel(logging.ERROR)
errorLogHandler.setFormatter(formatter)

logger.addHandler(logHandler)
logger.addHandler(errorLogHandler)

# YOUR MODULE ,WHERE YOU CAN PUT LOGS
def inference(model_name):
    start_time=time.time()
    #---------------
    # YOUR ALGORITHMS
    #---------------
    time.sleep(2)
    end_time=time.time()
    process_time=str(end_time-start_time)
    logger.info('model_name - '+model_name+',process_time - '+process_time)
    logger.error("An error log statement")
    logger.removeHandler(logHandler)
    logger.removeHandler(errorLogHandler)

if __name__=="__main__":
    inference("face_detection")

```

```python
"""
@Copyright : Cogknit Semantics pvt limited
@Author :    Suresh Saini
@Email  :    suresh@cogknit.com
@Date    :   12th April 2018
"""
import os
import time
import traceback
from django_redis import get_redis_connection
class RedisBroker(object):
    """
    Redis broker class
    """
    def __init__(self):
        # connect to redis server
        self.redis_conn=get_redis_connection("default")
    def get_redis_connection(self):
        return self.redis_conn

    def subscriber(self):
        try:
            p = self.redis_conn.pubsuab()  # See https://github.com/andymccurdy/redis-py/#publish--subscribe
            p.subscribe('startScripts')  # Subscribe to startScripts channel
            PAUSE = True
            start_new_task=False
            while PAUSE:  # Will stay in loop until START message received
                message = p.get_message()  # Checks for message
                if message:
                    command = message['data']  # Get data from message
                    if command == b'START':  # Checks for START message
                        PAUSE = False  # Breaks loop
                        start_new_task=True
                time.sleep(1)
            return start_new_task

        except Exception as e:
            print("!!!!!!!!!! EXCEPTION !!!!!!!!!")
            print(str(e))
            print(traceback.format_exc())
            return False


    def publisher(self):
        try:

            # HERE SOME INITIAL WORK IS DONE THAT SCRIPTS 1 & 2 NEED TO WAIT FOR
            # IDs SERIAL PORTS
            # SAVE TO db
            status=True
            if status:    
                self.redis_conn.publish('startScripts', 'START')  # PUBLISH START message on startScripts channe

        except Exception as e:
            print("!!!!!!!!!! EXCEPTION !!!!!!!!!")
            print(str(e))
            print(traceback.format_exc())

if __name__=="__main__":
    os.environ['DJANGO_SETTINGS_MODULE'] = 'acc_apigateway.settings'
    rb=RedisBroker()
    rb.publisher()
```
